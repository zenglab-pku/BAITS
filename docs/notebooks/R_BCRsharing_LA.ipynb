```r
library(dplyr)
library(purrr)
library(ggpubr)
library(stringr)
library(ggeasy)
library(tidyr)
library(ggplot2)
library(pheatmap)
library(org.Hs.eg.db)
suppressMessages(library(clusterProfiler))
suppressMessages(library(org.Hs.eg.db))
library(Seurat)
library(harmony)
```

```r
# 细胞类型和 Overlap 的相关性分析
celltype_comparison = c('Fibroblast','B','Macrophages','T','Endothelial','DC','NK')
T_subtype_comparison = c('Other T','Tfh','Treg','effector CD8T','exhausted CD8T','memory T','naive T')
B_subtype_comparison = c('Bn_TCL1A', 'ASC', 'GC_B', 'AtM', 'Bm', 'Bm_stress')
macro_subtype_comparison = c('C1Q+ macro', 'CXCL9+ macro', 'IL1+ macro', 'SPP1+ macro')
fibro_subtype_comparison = c( 'Progenitor-like', 'iCAF', 'myCAF') 
endo_subtype_comparison = c('Artery','Capillary','Lymphatic','Pericyte','Tip cell','Venule') 
bcr_comparison = c('count', 'expansion', 'clonality', 'gini', 'mutation_level', 'vmutation')
```

## Percent 影响进入肿瘤的因素

```r
cutoff = 10000
```

```r
overlap_0 = read.table('/home/zhaoyp/1.CRLM_XCR/2.XCR_again/files/2.LA/all_la_info.txt',sep = '\t',header = TRUE,check.names = FALSE)
overlap_0 = overlap_0[overlap_0$area >= cutoff,]
overlap_0[is.na(overlap_0)] = 0
```

```r
cellratio = read.csv('/home/zhaoyp/1.CRLM_XCR/2.XCR_again/files/2.LA/la_sptial_ct_Subtype2Allcells.csv', sep = ',', header = TRUE, check.names = FALSE)
```

```r
overlap_new = overlap_0[,c('sample','aggregates_id','patient','tissue','percent','normalized_percent','pathology_two','pathology_three', bcr_comparison)]
overlap_new = merge(overlap_new, cellratio, by=c('sample','aggregates_id'), all.x = TRUE)
overlap_new = overlap_new[overlap_new$sample != 'P1019-LM',] 
overlap = overlap_new

overlap = overlap[overlap$pathology_three %in% c('Normal-Tumor','Normal'), ]
```

```r
colnames(overlap)
```

```r
comparison = c('distance', celltype_comparison)

cor_results <- data.frame( comparison = character(), correlation = numeric(), p_value = numeric(), stringsAsFactors = FALSE) 
for (i in 1:length(comparison)) {
  x <- overlap[,'percent'][!is.na(overlap[,comparison[i]])]
  y <- overlap[,comparison[i]][!is.na(overlap[,comparison[i]])] 
  if(length(x) > 1 && length(y) > 1) {  
    test_result <- cor.test(x, y)
    cor_results <- rbind(cor_results, data.frame( comparison = comparison[i], correlation = test_result$estimate, p_value = test_result$p.value ))
  }
}

cor_results['Sig'] = ifelse(cor_results$p_value<0.05, 'Sig', 'No_Sig')
cor_results['color'] = ifelse(cor_results['correlation']>0, 'Sig. Positive', 'Sig. Negative') 
cor_results[cor_results$Sig=='No_Sig', 'color'] = 'No_Sig'
```

```r
options (repr.plot.width = 6, repr.plot.height = 3.5)
ggplot(cor_results, aes(x = reorder(comparison, correlation), y = correlation, fill = color)) +
  geom_bar(stat = "identity", width = 0.7) +
  scale_fill_manual(values = c("Sig. Negative" = "#3c52c2", "Sig. Positive" = "#bc2c17",'No_Sig'='grey')) +
  geom_text(aes(label = ifelse(p_value < 0.001, "***", ifelse(p_value < 0.01, "**", ifelse(p_value < 0.05, "*", "ns."))), y = max(cor_results$correlation)+0.05, size = 3.5)) + 
  # scale_fill_gradient2(low = "blue", mid = "white", high = "red", midpoint = 0, limits = c(-0.5, 0.5)) +
  labs(x = "", y = "Correlation Coefficient", title = "") +
  theme_minimal() +
  guides(fill = guide_legend(title = "")) +
  theme(axis.text.x = element_text(angle = 45, hjust = 1), plot.title = element_text(size=9, hjust = 0.5), axis.title = element_text(size = 9), #legend.position = "none",
        panel.grid = element_blank(), axis.line = element_line(color = "black")) +
  ylim(min(cor_results$correlation) - 0.1, max(cor_results$correlation) + 0.1) + 
  geom_vline(xintercept = 1.5, linetype = "dashed", color = "gray50") 

ggsave('/home/zhaoyp/1.CRLM_XCR/2.XCR_again/figures/4.fibro/sharedBCRs_LA_cts.pdf', width=5.5, height=2.5)
```
